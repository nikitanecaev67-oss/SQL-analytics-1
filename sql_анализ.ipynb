```python
options=csv_to_sqlite.CsvOptions(typing_style='full',encoding='utf-8')
input_name = ['targets.csv', 'reciepts.csv',
              'pastry_inventory.csv', 'staff.csv', 'sales_outlet.csv',
              'product.csv', 'customer.csv','generations.csv' ,'dates.csv' ]
database_name = 'output.sqlite'
if os.path.exists(database_name):
    os.remove(database_name)
csv_to_sqlite.write_csv(input_name, database_name, options)
```

```python
con=sqlite3.connect('output.sqlite')
```

```python
#найти всех женщин, у которых email заканчивается на gov
targets_1 = pd.read_sql_query('''SELECT *
                                    FROM customer
                                    WHERE customer_email like '%gov' AND gender = 'F'
                                    ''',con)
targets_1
```

```python
# найти сколько выручки принесли пользователи старше 30 лет
task_2 = pd.read_sql_query('''SELECT customer_id, sum(unit_price * quantity) total
                                    FROM reciepts
                                    inner join customer using(customer_id)
                                    where birthdate > 1995-01-01
                                    group by customer_id
                                    ''',con)
task_2
```

```python
#количество уникальных покупок за 2 последние недели апреля  (80 уникальных товаров)
task_2 = pd.read_sql_query('''SELECT DISTINCT product_id
                                    FROM reciepts
                                    WHERE transaction_date >= '2019-04-15'
                                    ''',con)
task_2
```
